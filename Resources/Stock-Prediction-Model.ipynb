## 4.Data Preprocessing

```python
df.shape
```

```python
# Splitting data into Training & Testing datasets
data_training = pd.DataFrame(df.Close[0:int(len(df)*0.7)])
data_testing = pd.DataFrame(df.Close[int(len(df)*0.7):int(len(df))])
```

```python
data_training
```

```python
df.shape
```

```python
#Scaling own the data between 0 and 1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
type(data_training_array)
```

```python
data_training_array.shape
```

## 5.Sequence Creation

```python
x_train = [] # previous 100 data
y_train = [] # new predicted data

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i, 0])
x_train, y_train = np.array(x_train), np.array(y_train) 
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
y_train.shape
```

## 6.Model Building

```python
import tensorflow as tf
print(tf.__version__)
```

```python
#ML Model
from keras.models import Sequential
from keras.layers import Dense, LSTM, Input
```

```python
model = Sequential() 

model.add(Input(shape=(100, 1))) # 100 - previous prices , 1- feature/ closing price
model.add(LSTM(units=128, activation='tanh', return_sequences=True)) # Units - 128 - 128 neurons, activation -- tanh - hyperbolic tangent - handles both (+)ve and (-)ve values, return_sequences=True -- will return sequence of outputs to the next layer
model.add(LSTM(units=64)) # -- By default -- return_sequences=False -- will generate and return single output, basically summary of 128 into 64
model.add(Dense(25)) # Dense layer -- typically used in the end of N.N to transform the learned features into the final output, will analyze it in 25 different angles , and finally pass it to the next layer will pass 25
model.add(Dense(1)) # Will take 25 outputs from previous Dense layer and combines them into a single output, {(1) neuron} , so it will give final prediction of price. | so, we have come from 100 inputs to single output |
```

## 7.Model Training

```python
model.compile(optimizer='adam', loss='mean_squared_error') # adam -> optimizing algo
model.fit(x_train, y_train, epochs=50)
```

```python
model.summary()
```

```python
# Save the model
model.save("stock_model.h5")  # Simple and backward-compatible
```

## 10. Model Evaluation

```python
# Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_predicted)
print(f"Mean Squared Error (MSE): {mse}")    # Lower the MSE better the predictions are
```

```python
# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Squared Error (MSE): {rmse}")  # Actual value only differs by this units
```

```python
# R-Squared 
r2 = r2_score(y_test, y_predicted)
print(f"R-Squared: {r2}") # Closer to 1 -- better the predictions are
```

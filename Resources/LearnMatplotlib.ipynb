```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Types of Plots

## Line Plots

```python
x = [1, 2, 3, 4, 5]
y = [10, 22, 20, 41, 60]
```

```python
plt.plot(x,y);
plt.title('Simple Line Plot');
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.grid(True)
plt.show()
```

## Bar Chart

```python
plt.figure(figsize=(7, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 210, 150, 300 ]
plt.bar(months, revenue)
plt.title('Monthly Revenue Bar Chart')
plt.xlabel('Months')
plt.ylabel('Revenue')
```

## Scatter Plot

```python
heights = [65, 70, 68, 72, 66]
weights = [150, 60, 155, 180, 145]
```

## Subplots & Legends

### Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 18, 25, 35, 35]
```

```python
plt.plot(days, original_price, 'g-o' , label='Original Price')
plt.plot(days, predicted_price, 'r-o' , label='Predicted Price')
plt.legend()
plt.title('Plot with Legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
```

### Subplots

```python
# First subplot - Original Price 
plt.subplot(3, 1, 3)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.title('Original prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Second subplot - Predicted Price 
plt.subplot(3, 1, 3)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Predicted prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Third subplot - Difference between Original & Predicted Prices
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3)
plt.bar(days, price_difference, color='purple', label='Price difference')
plt.title('Difference between Original & Predicted Prices')
plt.xlabel('Days')
plt.ylabel('Price Difference')
plt.legend()

plt.tight_layout()
```

#### Saving Figure

```python
plt.plot(days, original_price, 'g-o' , label='Original Price')
plt.plot(days, predicted_price, 'r-o' , label='Predicted Price')
plt.legend()
plt.title('Plot with Legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.savefig('Prediction_Image.png', dpi=300) # save the image
```
